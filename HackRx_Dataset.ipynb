```python
import random
def add():
    dict_customer_city = {}
    
    for i in customers:
        index = random.randint(0,7)
        dict_customer_city[i]=cities[index]
    return dict_customer_city 
dict_customer_city = add()
```

```python
def add1():
    dict_customer_seg = {}
    
    for i in customers:
        index = random.randint(0,4)
        dict_customer_seg[i]=segment[index]
    return dict_customer_seg
dict_customer_seg = add1()
```

```python
d1['City'] = d1['Customer'].apply(lambda x:dict_customer_city[x])
```

```python
d1['Segment'] = d1['Customer'].apply(lambda x:dict_customer_seg[x])
```

```python
import random
ratings_list = range(1,6)
product_ratings = random.choices(ratings_list,k=1500) 
```

```python
df = pd.read_csv(r'C:\Users\dp\Downloads\Customer_Acqusition.csv')
df
```

```python
age_array = list(df['Age'])
customer_array = list(df['Customer'])
result_dict = list(zip(customer_array,age_array))
dict_customer_age = {}
for i in result_dict:
    dict_customer_age[i[0]] = i[1]
```

```python
d1['Age'] = d1['Customer'].apply(lambda x:dict_customer_age[x])
d1
```

```python
d1['Product-ratings'] = product_ratings
d1
```

```python
gender_list = random.choices(['M','F'],k=100)
gender_list = list(zip(customer_array,gender_list))
gender_dict = {}
for i in gender_list:
    gender_dict[i[0]] = i[1]
d1['gender'] = d1['Customer'].apply(lambda x:gender_dict[x])
d1
```

```python
# Removing redundant features from dataframe.
df_final = d1.drop(['Sl No:', 'Amount'], axis = 1)

# Setting all under 18 status as Single
df_final['Relationship Status'] = [random.choices(['Single','Married'])[0] if df_final['Age'][i] > 18 else "Single" for i in range(len(df_final['Customer']))]

pd.set_option('max_colwidth', 100)
pd.set_option("display.max_rows", None, "display.max_columns", None)

# Converting Month column into Datetime
df_final['Month'] = df_final['Month'].apply(pd.to_datetime)
df_final
```
